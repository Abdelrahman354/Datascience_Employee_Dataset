```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.linear_model import LinearRegression
import os

# Configure TensorFlow to use GPU
os.environ['CUDA_VISIBLE_DEVICES'] = '0'
# Reading CSV file from specified path
data = pd.read_csv(r"C:\Users\abdel\Downloads\data\ProjectData.csv")

```

```python
# Function to fix 'Marital Status' column values

def fix_Marital(a):
    if(a=="M" or a=="m" or a=="married" or a=="Married"):
        return "married"
    else:
        return "single"
data["Marital Status"]=data["Marital Status"].apply(fix_Marital)

data[' Income '] = data[' Income '].str.replace('$', '', regex=False).str.replace(',', '', regex=False)


# convert income data type to numeric 
data[' Income '] = pd.to_numeric(data[' Income '], errors='coerce')

```

```python
# information about the dataframe before transformation
print(data.info())

# Function to standardize 'Children' column values
def fix_Children(a):

    if (a=="one"  or a=="1"):
        return "1"
    elif (a=="zero"  or a=="0"):
        return "0"
    elif (a=="two"  or a=="2"):
        return "2"
    elif (a=="three"  or a=="3"):
        return "3"
    elif (a=="four"  or a=="4"):
        return "4"
    elif (a=="five"  or a=="5"):
        return "5"
    elif (a=="six"  or a=="6"):
        return "6"
    else:
        return "0"
    
data['Children' ] = data['Children'].apply(fix_Children).astype(int)


```

```python

# Function to standardize 'Purchased Bike' column values
def fix_PurchasedBike(a):
    a_lower = a.lower().strip()  # Convert to lowercase and strip whitespace
    if a_lower in ["y", "yes"]:
        return 1
    else:
        return 0

    
data['Purchased Bike']=data['Purchased Bike'].apply(fix_PurchasedBike)



```

```python

# Function to standardize 'Home Owner' column values
def fix_home_owner(a):
    a_lower = a.lower().strip()  # Convert to lowercase and strip whitespace
    if a_lower in ["y", "yes"]:
        return 1
    else:
        return 0

# Applying the fix_home_owner function to 'Home Owner' column and converting to integer
data['Home Owner'] = data['Home Owner'].apply(fix_home_owner).astype(int)

# Function to standardize 'Gender' column values
def fix_gender(a):
    a_lower = a.lower().strip()  # Convert to lowercase and strip whitespace
    if a_lower in ["m", "male"]:
        return "male"
    else:
        return "female"

# Applying the fix_gender function to 'Gender' column
data['Gender'] = data['Gender'].apply(fix_gender)
```

```python
# Filling missing values with the mode (most common value) of each column
data = data.fillna(data.mode().iloc[0])

# Counting the number of duplicated rows in the dataframe
print("Number of duplicated data:", data.duplicated().sum(), "\n")

```

```python
# LabelEncoder
Encoder = LabelEncoder()

# convert columns to one hot encoding 
data['Gender'] = Encoder.fit_transform(data['Gender'])
data['Marital Status'] = Encoder.fit_transform(data['Marital Status'])
data['Education'] = Encoder.fit_transform(data['Education'])
data['Region'] = Encoder.fit_transform(data['Region'])

print(data.head(5))

```

```python

X=data[['Age','Gender', 'Marital Status', 'Education','Region','Children','Home Owner','Cars','Purchased Bike'] ]
y=data[' Income ']

selector = SelectKBest(score_func=f_regression, k=6)  # Select top 6 features based on F-test
X_selected = selector.fit_transform(X, y)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.33, random_state=44)

# Print the shape of X_selected after feature selection
print("Shape of X_selected after feature selection:", X_selected.shape)
```

```python

#the parameter grid for SVR
param_grid = {
    'C': [0.1, 1, 10, 100, 1000],
    'epsilon': [0.01, 0.1, 0.5, 1],
    'gamma': ['scale', 'auto'],
    'kernel':('linear','rbf')
}

# SVR model
svr = SVR()  

grid_search = GridSearchCV(estimator=svr, param_grid=param_grid, scoring='neg_mean_squared_error', cv=5)

grid_search.fit(X_train, y_train)

# Get the best SVR model from the grid search
best_svr = grid_search.best_estimator_

# predictions on the test set using the best SVR model
y_pred = best_svr.predict(X_test)

# Calculate RMSE (Root Mean Squared Error) on the test set
test_rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# Calculate MAE (mean_absolute_error) on the test set
test_mae = mean_absolute_error(y_test, y_pred)

print("Test MAE in  svr:",test_mae)

print("Test RMSE in svr:", test_rmse)
print("Best SVR Parameters:", grid_search.best_params_)
```

```python

print(y_pred[1],y_test[1])
y_test_values = y_test.values

# Plotting first 50 points
plt.plot(y_pred[:50], color='red', marker='o', linestyle='-', label='Predicted')  
plt.plot(y_test_values[:50], color='green', marker='o', linestyle='-', label='True')

# Add labels and title
plt.xlabel('Index (First 50)')
plt.ylabel('Value')
plt.title('True vs. Predicted Values (First 50)')

# Add legend
plt.legend()

# Show plot
plt.show()

```

```python
#Linear Regression model
LinearReg=LinearRegression(n_jobs=-1)

LinearReg.fit(X_train, y_train)


y_pred = LinearReg.predict(X_test)

# Calculate RMSE (Root Mean Squared Error) on the test set
test_rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# Calculate MAE (mean_absolute_error) on the test set
test_mae = mean_absolute_error(y_test, y_pred)

print("Test MAE in Linear Regression:",test_mae)
print("Test RMSE in Linear Regression:", test_rmse)
```

```python
from sklearn.tree  import DecisionTreeRegressor


#the parameter grid for DecisionTree Regressor
param_grid = {
    'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson']
    ,'max_depth':[5,10,25,50,100],
    'splitter':['best', 'random']
}

# DecisionTree Regressor model 
DecisionTreeRegressorr=DecisionTreeRegressor(random_state=42)

grid_search = GridSearchCV(estimator=DecisionTreeRegressorr, param_grid=param_grid, scoring='neg_mean_squared_error', cv=5)

grid_search.fit(X_train, y_train)

# Get the best DecisionTree Regressor model from the grid search
best_svr = grid_search.best_estimator_

# predictions on the test set using the best DecisionTree Regressor model
y_pred = best_svr.predict(X_test)

# Calculate RMSE (Root Mean Squared Error) on the test set
test_rmse = np.sqrt(mean_squared_error(y_test, y_pred))
# Calculate MAE (mean_absolute_error) on the test set
test_mae = mean_absolute_error(y_test, y_pred)

print("Test MAE in DecisionTreeRegressor:",test_mae)
print("Test RMSE in DecisionTreeRegressor:", test_rmse)
print("Best DecisionTree Regressor Parameters:", grid_search.best_params_)
```
